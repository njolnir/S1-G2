```python
# Load the data and Select PH data
data = pd.read_csv('0-Data/micro_world.csv', encoding='ISO-8859-1')
data_ph = data.query('economy == "Philippines"')
data_ph.info()
```

```python
# Load the data and Select Asean
Asean= ['Singapore', 'Thailand', 'Indonesia', 'Vietnam', 'Philippines', 'Malaysia', 'Cambodia', 'Brunei', 'Lao PDR', 'Myanmar']
data_asean = data.query('economy in @Asean')
data_asean['economy'].unique()
```

```python
#compare PH savers vs Asean Counterparts
data_asean = data.query('economy in @Asean')
data_asean.info()
```

```python
#groupby economy and get %SOA
Asean_SOA= data_Asean.groupby(['economy', 'economycode', 'regionwb']).agg(
    POP = ('wpid_random', 'count'),
    SOA = ('SOA', 'sum'))
Asean_SOA['SOA_percent'] = round(100 * Asean_SOA['SOA'] / Asean_SOA['POP'],2)
Asean = Asean_SOA.sort_values('SOA_percent', ascending=False).reset_index()
Asean
```

```python
# plot a bar plot
plt.figure(figsize=(4,3), dpi=320)
cols = ['blue' if x == "PHL" else 'lightgrey' for x in Asean['economycode']]
sns.set_style('ticks')
ax = sns.barplot(x="economycode", y="SOA_percent", data=Asean, palette=cols)

plt.xlabel('Country')
plt.title('Asean Country: %SOA')


```

```python
#Correlation of Ages vs Financial worry(old age) of those who did not saved for old age
SOA = data_ph.query('fin16 == 2')
SOA_FW = SOA[['economy', 'economycode','regionwb', 'fin16', 'fin44a', 'age']]
SOA_FW = SOA_FW[SOA_FW['fin44a'] < 4]
SOA_FW['fin44a'] = SOA_FW['fin44a'].replace({1: 'Very worried', 2: 'Somewhat worried', 3: 'Not worried'})
SOA_FW
```

```python
#Boxplot the SOA_FW
plt.figure(figsize=(3,4)  , dpi=320)
sns.set_style('whitegrid')
sns.boxplot(y='age', x='fin44a', data=SOA_FW, palette='Blues')
plt.ylabel('Age')
plt.xlabel('Worry Level')
plt.xticks(rotation=45)

plt.show()
```
